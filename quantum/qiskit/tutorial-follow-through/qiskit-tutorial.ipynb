# Learning Qiskit Follow Through

```python
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(3, 3)
qc.draw()
```

```python
qc.measure([0, 1, 2], [0, 1, 2])
qc.draw()
```

## Simulation

```python
from qiskit.providers.aer import AerSimulator
sim = AerSimulator()
```

```python
job = sim.run(qc)
result = job.result()
result.get_counts()
```

## Cirtuit with gates

```python
qc = QuantumCircuit(3, 3)
qc.x([0, 1])
qc.measure([0,1,2], [0,1,2])
qc.draw()
```

```python
job = sim.run(qc)
result = job.result()
result.get_counts()
```

## Addition circuit and CNOT

```python
qc = QuantumCircuit(2, 2)
qc.x(0)
qc.cx(0,1)
qc.measure([0,1], [0,1])
display(qc.draw())

job = sim.run(qc)
result = job.result()
print("Result: ", result.get_counts())
```

## Full half adder circuit

```python
test_qc = QuantumCircuit(4, 2)
test_qc.x(0)
test_qc.x(1)

test_qc.cx(0, 2)
test_qc.cx(1, 2)
test_qc.ccx(0,1,3)

test_qc.measure(2,0)
test_qc.measure(3,1)
test_qc.draw()
```

```python
job = sim.run(test_qc)
result = job.result()
result.get_counts()
```
