```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(3, 3)

qc.measure([0,1,2],[0,1,2]) # Measure quantum bit into classical bits
qc.draw()
```

## Simulation

```python
from qiskit.providers.aer import AerSimulator
sim = AerSimulator()
```

```python
job = sim.run(qc)
result = job.result()
result.get_counts()
```

### Product States

```python
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.h(1)

qc.measure([0,1],[0,1]) # Measure quantum bit into classical bits
qc.draw()
```

```python
job = sim.run(qc)
result = job.result()
result.get_counts()
```

### Entangled States

```python
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0,1)
qc.measure([0,1],[0,1]) # Measure quantum bit into classical bits
qc.draw()
```

```python
job = sim.run(qc)
result = job.result()
result.get_counts()
```

### Multiple measurements

```python
qc = QuantumCircuit(1,4)
qc.h(0)
qc.measure([0], [0])
qc.y(0)
qc.measure([0], [1])
qc.y(0)
qc.measure([0], [2])
qc.y(0)
qc.measure([0], [3])
qc.draw()
```

```python
job = sim.run(qc)
result = job.result()
result.get_counts()
```

### Phase operator

```python
from math import pi

qc = QuantumCircuit(1,1)
qc.ry(-pi/4, 0)
qc.measure([0],[0])
qc.draw()
```

```python
job = sim.run(qc)
result = job.result()
result.get_counts()
```
